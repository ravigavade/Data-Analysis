Accessing some specific data

 --we us loc and iloc

loc:
  - allows us to filter by rows and columns with names


iloc:
  - allows us to filter by rows n colums with indices

```python
coffee.head()
```

```python
coffee.loc[0]
```

```python
coffee.loc[[0,1,2]]
```

```python
coffee.loc[2:4]
```

```python
coffee.loc[2,"Units Sold"]
```

```python
coffee.loc[0:4,["Coffee Type","Units Sold"]]
```

```python
coffee.iloc[0:4,[1,2]]
```

```python
#chaning the value
coffee.loc[1,"Units sold"]=20
```
